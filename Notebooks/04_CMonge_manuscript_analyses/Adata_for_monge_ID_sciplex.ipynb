```python
import scanpy as sc
from sklearn.model_selection import train_test_split
```

```python
with open("/path/to/data/datasets/scrna-sciplex3/drugs.txt", "r") as f:
    drugs = f.readlines()
drugs = [s.strip() for s in drugs]
drugs
```

```python
adata = sc.read_h5ad("/path/to/data/datasets/scrna-sciplex3/hvg.h5ad")
sel_adata = adata[adata.obs["drug"].isin(drugs+["control"])]
```

```python
sel_adata.obs["drug"].unique()
```

```python
sel_adata.obs
```

```python
# One model for all drugs
# Make train-test split here to make training with all drugs and evaluation per drug possible
train, test = train_test_split(sel_adata.obs.index, stratify=sel_adata.obs["drug"], train_size=0.8)
sel_adata.obs["monge-id"] = None
sel_adata.obs.loc[sel_adata.obs_names.isin(train), "monge-id"] = "train"
sel_adata.obs.loc[sel_adata.obs_names.isin(test), "monge-id"] = sel_adata.obs.loc[sel_adata.obs_names.isin(test), "drug-dose"]

sel_adata.obs.loc[(sel_adata.obs["monge-id"]=="train")&(sel_adata.obs["drug"]=="control"), "monge-id"] = "train_control"
```
